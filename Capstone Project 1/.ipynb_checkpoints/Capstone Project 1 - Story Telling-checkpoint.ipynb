### Preliminary Actions

```python
# There are many warnings regarding the updates in the future releases of the libraries. Ignore them.
import warnings
warnings.filterwarnings("ignore")

# Set the backend of matplotlib to inline
%matplotlib inline

# Import necessary libraries
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Set style and context
sns.set_style('whitegrid')
sns.set_context('talk')

# Read the csv file created in the Data Wrangling section
df=pd.read_csv('wrangled_data.csv')

#Print the first five rows of the data
df.head()
```

### Step 1: Create the boxplot of all quantitative variables to understand their distribution based on the five number summary

```python
plt.figure(figsize=(15,10))
sns.boxplot(x=df.iloc[:,4:])
plt.title('Box plot of the four levels of education, poverty, and unemployment across US counties',fontsize=20)
plt.ylabel('percentage')
plt.xticks(rotation=30)
plt.show()
```

### Observations 
#### 1) Among the four education levels, 
####   a) less than high school and high school diploma respectively have the lowest and highest first quartile, median, and third quartile. College/associate degree and bachelors/higher are ranked second and third, respectively. 
####  b) College/associate degree has the lowest width of inter quartile range. It means it has the most dense distribution. On the other hand, high school diploma and bachelors/higher have the widest inter quartile range, but bachelors/higher have many more outliers than high school diploma, and so it has the least dense distribution

#### 2) Less than high school, bachelors/higher, poverty, and unemployment are skewed right because they have many outliers above the upper whisker. High school diploma is slightly skewed left. College/associate degree is almost not skewed

#### 3) Unemployment has the most dense distribution among all the quantitative variables. The distance between lower and upper whiskers is less than ten percent

### Step 2: Examine the relationship between each level of education on the one hand, and unemployment and poverty on the other hand
</div>

```python
# Change the plot style and font scale to have a clearer plot
sns.set(style="ticks", color_codes=True, font_scale=1.2)

# Create a dataframe with only six columns: four levels of education, unemployment, and poverty
df_reduced = df.loc[:,['less_than_high_school','high_school_diploma','college/associate_degree'\
                       ,'bachelors/higher','unemployment','poverty']]

# pairplot poverty and unemployment vs each level of education  and plot the regression line
g=sns.pairplot(df,x_vars=['less_than_high_school','high_school_diploma','college/associate_degree'\
                                ,'bachelors/higher'],y_vars=['unemployment','poverty'],kind="reg",markers='.')
g.fig.suptitle("Pairplot of unemployment and poverty vs. the four education levels across US counties",fontsize=20)

# Create the correlation matrix and separate the target elements
corr_df_reduced = df_reduced.corr().iloc[4:,:4]

# Resize the plot for heatmap
plt.figure(figsize = (15,5))
# Plot the heatmap, annotate it, and resize the annotation
sns.heatmap(corr_df_reduced,annot=True,annot_kws={"size": 20})
plt.title('Heatmap of unemployment and poverty vs. the four education levels across US counties',fontsize=20)
plt.show()
```

### Observation 
#### As the education level goes higher, the slope of regression line as well as the correlation coefficient between the education level and both poverty and unemployment go lower. In other words, higher education results in lower unemployment and poverty.

### Step 3: Examine the correlation between education, unemployment, and poverty all together. In order to simplify the interpretation, a new education variable is introduced to be used instead of the four education levels. This variable is percentage of the county population with some degree more than high school diploma. 

```python
# Define the new column for the new variable
df['more_than_high_school'] = df['college/associate_degree']+df['bachelors/higher']

# Create a dataframe with only three columns: more than high school, unemployment, and poverty
df_reduced = df.loc[:,['more_than_high_school','unemployment','poverty']]

#Create the paiplot between the three variables and add the regression line
g=sns.pairplot(df_reduced,kind="reg")
g.fig.suptitle("Pairplot of unemployment, poverty, and education across US counties",fontsize=20)

# Create the correlation matrix
corr_df_reduced = df_reduced.corr()

#Resize the plot for heatmap
plt.figure(figsize = (8,5))

# Plot the heatmap, annotate it, and resize the annotation
sns.heatmap(corr_df_reduced,annot=True,annot_kws={"size": 20})
plt.title('Heatmap of unemployment, poverty, and education across US counties',fontsize=20)
plt.show()
```

### Observation 
#### Unemployment and poverty are strongly correlated to each other. In addition, both have strong negative correlation with education

### Step 4: Create the boxplot of the four education levels as well as unemployment and poverty across the counties in each of the nine RUCC. Two box plots are created to avoid a single busy plot

```python
# Set style
sns.set_style('whitegrid')

# Sort the RUCC column
df.sort_values(['RUCC'],inplace=True)

#Melt the dataframe based on the four education levels
df_melt1 = pd.melt(df,id_vars='RUCC',value_vars=['less_than_high_school','high_school_diploma','college/associate_degree',\
                                                'bachelors/higher'], var_name='variable',value_name='percentage')

#Melt the dataframe based on unemployment and poverty
df_melt2 = pd.melt(df,id_vars='RUCC',value_vars=['unemployment','poverty'], var_name='variable',value_name='percentage')

# Create a two row one column subplot
figure, (ax1,ax2) = plt.subplots(2,1,figsize=(20,15))

# Create the boxplot of the four education level for each RUCC code
sns.boxplot(ax=ax1, x='RUCC',y='percentage',hue='variable',data=df_melt1,fliersize=2,width=0.6)
ax1.legend(loc='lower left',bbox_to_anchor=(0.4,0.73),fontsize=14)
ax1.set_ylim(0,70)
ax1.set_title('Box plot of the four education levels of counties in each of the nine RUCC',fontsize = 20)
ax1.set_xlabel('RUCC',fontsize = 20)
ax1.set_ylabel('percentage',fontsize = 20)

# Create the boxplot of the poverty and unemployment for each RUCC code
sns.boxplot(ax=ax2, x='RUCC',y='percentage',hue='variable',data=df_melt2,fliersize=2,width=0.6)
ax2.legend(loc='lower left',bbox_to_anchor=(0.4,0.8),fontsize=20)
ax2.set_title('Box plot of unemployment and poverty of counties in each of the nine RUCC',fontsize = 20)
ax2.set_xlabel('RUCC',fontsize = 20)
ax2.set_ylabel('percentage',fontsize = 20)

#Adjust the distance between the two subplots
plt.subplots_adjust(hspace=0.3)
plt.show()
```

### Observations
#### 1) Regarding the four education levels, 
#### a) the major difference across the nine RUCC is the distribution of bachelors/higher across metropolitan counties, i.e. RUCC 1, 2, and 3 on the one hand, and the non-metropolitan counties ,i.e. RUCC 4 to 9, on the other hand; the bachelors/higher has a higher median and wider inter quartile range in the metropolitan counties (especially RUCC 1 which is counties in metro areas of 1 million population or more) compared to the non-metropolitan counties
#### b) As it was observed before, less than high school, bachelors/higher, college/associate degree, and high school diploma have lowest to highest median, first and third quartiles over the entire US. This pattern is not observed in RUCC 1 since the median of bachelors/higher is close to the medians of high school diploma and college/associate degree. Its third quartile is even larger than theirs. As the plot moves from metropolitan counties toward the non-metropolitan counties, the box plot of bachelors/higher gets closer to the box plot of less than bachelors and further from those of the other two education levels. This results in the distributions of the four education levels of RUCC 2-5 approximately follow the general trend, but in higher RUCC, compared to the general trend, the box plot of bachelors/higher is too close to the box plot of less than high school diploma, i.e. their median, first and third quartiles are close to each other.

#### 2) The first quartile, median, and third quartile of the poverty in RUCC 1 is significantly lower compared to the rest of RUCC. This matches the observations from steps 2 and 3 that higher education results in lower poverty.

#### 3) There is not a major difference between the unemployment distribution of the nine RUCC.

### Step 5: Create the barplot of the mean of each six quantitative variables of counties across the nine RUCC

```python
# Melt the dataframe across all quantitative variables 
df_melt3 = pd.melt(df,id_vars='RUCC',value_vars=['less_than_high_school','high_school_diploma',\
                                                 'college/associate_degree','bachelors/higher','unemployment',\
                                                 'poverty'], var_name='variable',value_name='percentage')

# Create the bar plot
plt.figure(figsize=(15,10))
sns.barplot(x='variable',y='percentage',hue='RUCC',data=df_melt3,palette='RdBu',ci=None)
plt.title('Bar plot of the mean of each education level, unemployment, and poverty of counties across nine RUCC',fontsize = 20)
plt.legend(fontsize=20)
plt.xlabel('variable',fontsize = 20)
plt.ylabel('percentage',fontsize = 20)
plt.tick_params(axis='x', labelsize=12)
plt.show()
```

### Observation
#### As the plot moves from metropolitan counties or lower RUCC to non-metropolitan counties or higher RUCC, approximately:
#### 1) The mean of less than high school and the mean of high school diploma go up while the mean of bachelors/higher goes down. The average of some college or associate degree almost remains constant.
#### 2) The average of poverty goes up. Again this matches the observation that higher education results in lower poverty
#### 3) The average of unemployment remains constant

### Step 6: Examine the four levels of education, unemployment, and poverty across the counties four geographical regions; south, west, northeast, and midwest

```python
# Melt the dataframe across all quantitative variables 
df_melt4 = pd.melt(df,id_vars='region',value_vars=['less_than_high_school','high_school_diploma',\
                                                 'college/associate_degree','bachelors/higher','unemployment',\
                                                 'poverty'], var_name='variable',value_name='percentage')

# Create a two row one column subplot
figure, (ax1,ax2) = plt.subplots(2,1,figsize=(20,15))

# Create the boxplot of the four education level based on RUCC code
sns.boxplot(ax=ax1, x='region',y='percentage',hue='variable',data=df_melt4)
ax1.set_title('Box plot of the four education levels of counties across the four geographical regions of US',fontsize = 20)
ax1.set_xlabel('region',fontsize = 20)
ax1.set_ylabel('percentage',fontsize = 20)

# Create the bar plot
sns.barplot(x='variable',y='percentage',hue='region',data=df_melt4,palette='RdBu',ci=None)
plt.title('Bar plot of the average of each education level, unemployment, and poverty of counties across the\
 four geographical regions of US',fontsize = 20)
plt.legend(fontsize=20)
plt.xlabel('variable',fontsize = 20)
plt.ylabel('percentage',fontsize = 20)
plt.tick_params(axis='x', labelsize=12)
plt.show()
```

### Observation
#### 1) South has the highest less than high school and lowest bachelors/higher. This explains its highest poverty rate
#### 2) Northeast has higher bachelors/higher than midwest and west but lower college/associate degree than theirs. The poverty rates of the three regions are close to each other
